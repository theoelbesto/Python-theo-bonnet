```python
def gauss(pts,x0,xmin,xmax,s):
    x = np.linspace(xmin, xmax, pts)

    y = np.exp(-0.5 * ((x - x0) / s) ** 2) / (s * np.sqrt(2 * np.pi))

    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('Gauss')
    plt.title('Courbe de Gauss')
    plt.grid(True)

    plt.show()
gauss(100,0,-1,1,0.3)
```

```python
def Comp(pts):
    x = np.linspace(0, 3, pts)

    y = np.exp(-x)
    plt.plot(x, y,label="exp(-x)")
    plt.legend()
    
    y = np.exp(-x)*np.sin(3*np.pi*x)

    plt.plot(x, y,label='exp(-x)*sin(3*pi*x)')
    plt.xlabel('x')
    plt.ylabel('Y')
    plt.legend()
    plt.title('Comparaison')
    plt.grid(True)

    plt.show()
Comp(100)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# Generate synthetic data
X, y = datasets.make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create an SVM classifier with a linear kernel
svm_classifier = SVC(kernel='linear')

# Fit the SVM model to the training data
svm_classifier.fit(X_train, y_train)

# Make predictions on the test data
predictions = svm_classifier.predict(X_test)

# Calculate the accuracy of the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# Plot the decision boundary
def plot_decision_boundary(X, y, model):
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, s=30)
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
    Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

plt.figure(figsize=(8, 6))
plot_decision_boundary(X_test, y_test, svm_classifier)
plt.title("SVM with Linear Kernel")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# Generate synthetic data with non-linear separability
X, y = datasets.make_circles(n_samples=100, factor=0.5, noise=0.1, random_state=42)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create an SVM classifier with an RBF kernel
svm_classifier = SVC(kernel='rbf', gamma='scale')

# Fit the SVM model to the training data
svm_classifier.fit(X_train, y_train)

# Make predictions on the test data
predictions = svm_classifier.predict(X_test)

# Calculate the accuracy of the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

# Plot the decision boundary
def plot_decision_boundary(X, y, model):
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, s=30)
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
    Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])

plt.figure(figsize=(8, 6))
plot_decision_boundary(X_test, y_test, svm_classifier)
plt.title("SVM with RBF Kernel (Circular Decision Boundary)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mpl_toolkits.mplot3d import Axes3D

# Generate synthetic 3D data with two classes
np.random.seed(0)
X = np.random.randn(100, 3)
y = (X[:, 0] + X[:, 1] + X[:, 2] > 0).astype(int)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create an SVM classifier with an RBF kernel
svm_classifier = SVC(kernel='rbf', gamma='scale')

# Fit the SVM model to the training data
svm_classifier.fit(X_train, y_train)

# 3D Visualization of the decision boundary
def plot_3d_decision_boundary(X, y, model):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Scatter plot of data points
    ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.Paired, s=30)
    
    # Define a meshgrid in 3D space
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    z_min, z_max = X[:, 2].min() - 1, X[:, 2].max() + 1
    
    xx, yy, zz = np.meshgrid(np.linspace(x_min, x_max, 50), np.linspace(y_min, y_max, 50), np.linspace(z_min, z_max, 50))
    
    # Calculate decision function values for the meshgrid
    grid = np.c_[xx.ravel(), yy.ravel(), zz.ravel()]
    decision_values = model.decision_function(grid)
    
    # Reshape decision function values for plotting
    decision_surface = decision_values.reshape(xx.shape)
    
    # Plot the decision boundary as points where the decision value is zero
    decision_boundary = np.array(grid[np.abs(decision_values) < 0.1])
    ax.scatter(decision_boundary[:, 0], decision_boundary[:, 1], decision_boundary[:, 2], color='r', s=0.002)
    
    plt.show()

# Plot the 3D decision boundary
plot_3d_decision_boundary(X_test, y_test, svm_classifier)
```
