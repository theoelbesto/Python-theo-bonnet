```python
import numpy as np
import matplotlib.pyplot as plt

num = int(input("Enter the number of points to draw: "))

x = np.linspace(-1, 1, num)

y = np.cos(2 * np.pi * x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x) = cos(2*pi*x)')
plt.title('Graph of cos(2*pi*x)')
plt.grid(True)

plt.savefig('cos2pi.png')

plt.show()
```

```python
def gauss(pts,x0,xmin,xmax,s):
    x = np.linspace(xmin, xmax, pts)

    y = np.exp(-0.5 * ((x - x0) / s) ** 2) / (s * np.sqrt(2 * np.pi))

    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('Gauss')
    plt.title('Courbe de Gauss')
    plt.grid(True)

    plt.show()
gauss(100,0,-1,1,0.3)
```

```python
def Comp(pts):
    x = np.linspace(0, 3, pts)

    y = np.exp(-x)
    plt.plot(x, y,label="exp(-x)")
    plt.legend()
    
    y = np.exp(-x)*np.sin(3*np.pi*x)

    plt.plot(x, y,label='exp(-x)*sin(3*pi*x)')
    plt.xlabel('x')
    plt.ylabel('Y')
    plt.legend()
    plt.title('Comparaison')
    plt.grid(True)

    plt.show()
Comp(100)
```
